## MINI RAPPORT:

A travers la phase 3, nous devons créer des requêtes SQL afin d'interroger la base de données et visualiser les résultats.
On va donc vous expliquer chacune des requetes demandées.

Pour la requete 1, on nous demande "Combien de classes de passagers différentes y avait-il à bord du Titanic ?".
Ainsi pour y répondre, on a utilisé la fonction "select count" qui permet de compter et on l'a appliqué sur la table titanic afin que la requete nous affiche le nombre de classes différentes présentes.

Pour la requete 2,on nous demande "Combien de passagers y avait-il dans chaque classe?".
On va donc employer la fonction "group by" et "select count" qui permettent donc de projeter le nombre de passagers en fonction d'un ou plusieurs groupes (ici le groupe en question est la classe). 

Pour la requete 3, on nous demande "Combien de femmes et d'hommes y avait-il dans chaque classe?".
On va donc réemployer la fonction "group by" et "select count" comme à la requete précédente. Cependant, au lieu de se servir d'un seul groupe, on va en utiliser deux qui sont "le sexe" et "la classe" dans le but de projeter le nombre de passagers en fonction de chaque classe mais également du sexe.

Pour la requete 4, on nous demande "Comptez le nombre et le pourcentage de survivants et de passagers
morts."
On va donc utiliser le "select count" et le "where survived='0' ou '1'" afin de projeter uniquement le nombre de personnes mortes (ou vivantes) puis, on va diviser le nombre de personnes mortes ou vivantes (multiplié par 100) avec le nombre total de passagers pour projeter le pourcentage de personnes mortes (ou vivantes). On peut également renommer le nom des colonnes pour plus de précision (comme par exemple, percentalive pour le pourcentage de personnes vivantes).

Pour la requete 5, on nous demande "Visualiser la répartition des passagers survivants et morts par classe". Le terme "visualiser" nous indique qu'il faut donc créer une vue qui est une sorte de mini table que l'on peut afficher à tout moment avec l'expression "select * from nomview;". Ainsi on va utiliser la fonction "create view", "select count", "group by" (sur 1 groupe qui est la classe) et "where survived='0'(pour les morts) ou '1'(pour les vivants) afin de nous créer deux vues qui nous projeteront le nombre de passagers survivants décédés ou vivants en fonction de la classe. La création de ces vues qui porteront le nom de "nbmortsparclasse" pour celle qui affichera les morts par classe et "nbvivantsparclasse" pour les survivants nous permettront d'y avoir accès à tout moment comme avec une table en employant la commande "select * from nomview;"

Pour la requete 6, on nous demande "Visualiser la répartition des passagers survivants et des passagers
décédés selon le sexe".
Pour cette dernière requete, on fera la meme chose que la requete précédente sauf qu'au lieu d'utiliser le group by sur une colonne, on l'emploiera sur deux colonnes qui sont la 'classe' et le 'sexe' afin de nous créer deux vues qui nous projeteront le nombre de passagers survivants décédés ou vivants en fonction de la classe et du sexe. La création de ces vues qui porteront le nom de "nbmortsparclasseetsex" pour celle qui affichera les morts par classe et sex et "nbvivantsparclasseetsex" pour les survivants nous permettront d'y avoir accès à tout moment comme avec une table en employant la commande "select * from nomview;"



### LES REQUÊTES SQL

#### Requête 1:

```python
%%sql
select count(distinct Pclass) from titanic;
```

#### Requête 2:

```python
%%sql
select count(distinct name),Pclass from titanic group by Pclass;
```

#### Requete 3:

```python
%%sql
select count(distinct name),sex,Pclass from titanic group by Pclass,sex;
```

#### Requête 4.1:

```python
%%sql
select survived, count(*) as nbmorts, round(count(*) * 100/891) as percentdead from titanic where survived = '0' group by survived;
```

#### Requête 4.2:

```python
%%sql
select survived, count(*) as nbvivants, round(count(*) * 100/891) as percentalive from titanic where survived = '1' group by survived;
```

#### Requête 5.1:

```python
%%sql
DROP view nbmortsparclasse;
```

```python
%%sql
create view nbmortsparclasse as select count(distinct name) as mort, Pclass from titanic where survived = 0 group by Pclass;
select * from nbmortsparclasse;
```

#### Requête 5.2:

```python
%%sql
DROP view nbvivantsparclasse;
```

```python
%%sql
create view nbvivantsparclasse as select count(distinct name) as vivant, Pclass from titanic where survived = 1 group by Pclass;
select * from nbvivantsparclasse;
```

#### Requête 6.1:

```python
%%sql
DROP view nbvivantsparclasseetsex;
create view nbvivantsparclasseetsex as select count(distinct name) as vivant, Pclass, sex from titanic where survived = 1 group by Pclass,sex;
select * from nbvivantsparclasseetsex;
```

#### Requête 6.2:

```python
%%sql
DROP view nbmortsparclasseetsex;
create view nbmortsparclasseetsex as select count(distinct name) as mort, Pclass, sex from titanic where survived = 0 group by Pclass,sex;
select * from nbmortsparclasseetsex;
```
